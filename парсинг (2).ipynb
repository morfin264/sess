```python
el = 'https://www.dotabuff.com/'
x = []
for element in asa:
    
    result = el + element['href']
    x.append(result)
    print(result)
```

```python
requests.get(x[0],headers = {'User-agent': 'your bot 0.1'}).text
# BeautifulSoup(,'html.parser')
```

```python
name1 = []
role1 = []
for element in x:
    data=requests.get(element, headers = {'User-agent': 'your bot 0.1'})
    dt = BeautifulSoup(data.text, 'html.parser')
    name_desc = str(dt.find('h1'))
    name=name_desc[4:name_desc.find('<small>')]
    # print(name_desc[4:name_desc.find('<small>')])
    role=name_desc[name_desc.find('<small>')+7:name_desc.find('</small>')]
    # print(name_desc[name_desc.find('<small>')+7:name_desc.find('</small>')])
    #print(name_desc.find('<small>'))
    name1.append(name)
    print(name)
    role1.append(role)
    print(role)
```
