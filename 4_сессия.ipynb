```python
import re
import pandas as pd
import tkinter as tk
from tkinter import scrolledtext

# Пример данных (замените его своими реальными данными)

df = pd.DataFrame(data)

# Пример команд и их действий
commands = {
    "определить_адресата": {"Краткое содежание": None},
    "выдать_документы": {"дата регистрации": None, "Адресат": None},
    "справка": None,
}

def find_recipient_by_description(description):
    # Ваш код для определения адресата по описанию
    # (в данном примере просто возвращает первого адресата)
    return df[df["Краткое содежание"] == description]["Адресат"].values[0]

def issue_documents(period, recipient):
    # Ваш код для выдачи документов
    # (в данном примере просто возвращает список документов)
    return df[(df["дата регистрации"] == period) & (df["Адресат"] == recipient)]["Краткое содежание"].tolist()

def execute_command(command, params):
    if command == "определить_адресата":
        description = params["Краткое содежание"]
        try:
            recipient = find_recipient_by_description(description)
            return f"Определен 	Адресат: {recipient}"
        except IndexError:
            return "Адресат не найден"
    
    elif command == "выдать_документы":
        period = params["дата регистрации"]
        recipient = params["Адресат"]
        documents = issue_documents(period, recipient)
        if documents:
            return f"Выданы документы за период {period} для адресата {recipient}: {', '.join(documents)}"
        else:
            return f"Документы не найдены за период {period} для адресата {recipient}"
    
    elif command == "справка":
        return "Справка по командам: определить_адресата, выдать_документы, справка"
    
    else:
        return "Неизвестная команда"

def process_user_input():
    user_input = input_entry.get()
    if user_input.lower() == "выход":
        root.destroy()
    else:
        command = parse_command(user_input)
        if command:
            params = extract_params(user_input, command)
            result = execute_command(command, params)
            output_text.insert(tk.END, f"\nПользователь: {user_input}\nБот: {result}\n")
            output_text.yview(tk.END)
        else:
            output_text.insert(tk.END, f"\nПользователь: {user_input}\nБот: Не удалось распознать команду\n")
            output_text.yview(tk.END)

def parse_command(text):
    # Пример парсинга команды по ключевым словам с использованием регулярных выражений
    if "определить адресата" in text:
        return "определить_адресата"
    elif "выдать документы" in text:
        return "выдать_документы"
    elif "справка" in text:
        return "справка"
    else:
        return None

def extract_params(text, command):
    # Пример извлечения параметров из текста команды
    params = {}
    if command == "определить_адресата":
        match = re.search(r"определить адресата (.+)", text)
        if match:
            params["Краткое содежание"] = match.group(1)
    elif command == "выдать_документы":
        match = re.search(r"выдать документы (дата регистрации .+ Адресат .+)", text)
        if match:
            param_str = match.group(1)
            param_list = param_str.split()
            params["дата регистрации"] = param_list[param_list.index("дата регистрации") + 1]
            params["Адресат"] = param_list[param_list.index("Адресат") + 1]
    return params

# Создание графического интерфейса с использованием Tkinter
root = tk.Tk()
root.title("Бот")

# Создание текстового поля для вывода результата
output_text = scrolledtext.ScrolledText(root, height=20, width=50)
output_text.pack()

# Создание поля для ввода команды
input_entry = tk.Entry(root, width=50)
input_entry.pack()

# Создание кнопки "Отправить"
send_button = tk.Button(root, text="Отправить", command=process_user_input)
send_button.pack()

root.mainloop()
```
