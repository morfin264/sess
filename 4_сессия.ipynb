```python
import re
import pandas as pd
import tkinter as tk
from tkinter import scrolledtext

#   (    )

df = pd.DataFrame(data)

#     
commands = {
    "_": {" ": None},
    "_": {" ": None, "": None},
    "": None,
}

def find_recipient_by_description(description):
    #       
    # (      )
    return df[df[" "] == description][""].values[0]

def issue_documents(period, recipient):
    #     
    # (      )
    return df[(df[" "] == period) & (df[""] == recipient)][" "].tolist()

def execute_command(command, params):
    if command == "_":
        description = params[" "]
        try:
            recipient = find_recipient_by_description(description)
            return f" 	: {recipient}"
        except IndexError:
            return "  "
    
    elif command == "_":
        period = params[" "]
        recipient = params[""]
        documents = issue_documents(period, recipient)
        if documents:
            return f"    {period}   {recipient}: {', '.join(documents)}"
        else:
            return f"     {period}   {recipient}"
    
    elif command == "":
        return "  : _, _, "
    
    else:
        return " "

def process_user_input():
    user_input = input_entry.get()
    if user_input.lower() == "":
        root.destroy()
    else:
        command = parse_command(user_input)
        if command:
            params = extract_params(user_input, command)
            result = execute_command(command, params)
            output_text.insert(tk.END, f"\n: {user_input}\n: {result}\n")
            output_text.yview(tk.END)
        else:
            output_text.insert(tk.END, f"\n: {user_input}\n:    \n")
            output_text.yview(tk.END)

def parse_command(text):
    #          
    if " " in text:
        return "_"
    elif " " in text:
        return "_"
    elif "" in text:
        return ""
    else:
        return None

def extract_params(text, command):
    #      
    params = {}
    if command == "_":
        match = re.search(r"  (.+)", text)
        if match:
            params[" "] = match.group(1)
    elif command == "_":
        match = re.search(r"  (  .+  .+)", text)
        if match:
            param_str = match.group(1)
            param_list = param_str.split()
            params[" "] = param_list[param_list.index(" ") + 1]
            params[""] = param_list[param_list.index("") + 1]
    return params

#      Tkinter
root = tk.Tk()
root.title("")

#      
output_text = scrolledtext.ScrolledText(root, height=20, width=50)
output_text.pack()

#     
input_entry = tk.Entry(root, width=50)
input_entry.pack()

#   ""
send_button = tk.Button(root, text="", command=process_user_input)
send_button.pack()

root.mainloop()
```
